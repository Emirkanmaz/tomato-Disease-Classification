```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
BATCH_SIZE = 32
IMAGE_SIZE = 256
CHANNELS=3
```

# Data Augmentation. Used split_folders to split the dataset for train test validation. 

## split_folders -- output dataset -- ratio .7 .1 .2 -- dataset

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Veri artırma ayarlarını belirle
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

train_generator = train_datagen.flow_from_directory(
        'dataset/output/train',
        target_size=(IMAGE_SIZE,IMAGE_SIZE),
        batch_size=BATCH_SIZE,
        class_mode="sparse",
)

i=0
for image_batch, label_batch in train_generator:
    plt.figure(i)
    imgplot = plt.imshow(image_batch[0])
    i += 1
    if i % 4 == 0:
        break

plt.show()
```

```python
validation_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
validation_generator = validation_datagen.flow_from_directory(
        'dataset/output/val/',
        target_size=(IMAGE_SIZE,IMAGE_SIZE),
        batch_size=BATCH_SIZE,
        class_mode="sparse"
)
```

```python
test_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
test_generator = test_datagen.flow_from_directory(
        'dataset/output/test/',
        target_size=(IMAGE_SIZE,IMAGE_SIZE),
        batch_size=BATCH_SIZE,
        class_mode="sparse"
)
```

```python
n_classes = len(train_generator.class_indices.keys())
n_classes
```

```python
input_shape = (IMAGE_SIZE,IMAGE_SIZE,CHANNELS)

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, kernel_size=(3,3), activation="relu", input_shape=input_shape),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(128, kernel_size=(3,3), activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(n_classes, activation="softmax")    
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

```python
# total number of images / batch size is steps per epoch
12804/BATCH_SIZE
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=400,
    batch_size=32,
    validation_data=validation_generator,
    validation_steps=50,
    verbose=1,
    epochs=10,
)
```

```python
model.save("../tomato.h5")
```

```python
import numpy as np

class_names = list(test_generator.class_indices.keys())
i=0
plt.figure(figsize=(15,15))
for image_batch, label_batch in test_generator:
    
    image = image_batch[0]
    label = int(label_batch[0])
    plt.subplot(3, 3, i + 1)
    
    actual_class = class_names[label]
    
    prediction = model.predict(image_batch)
    
    predicted_class = class_names[np.argmax(prediction[0])]
    confidence = round(100 * (np.max(prediction[0])), 2)
    plt.axis("off")
    
    plt.title(f"Actual: {actual_class},\n Predicted: {predicted_class}.\n Confidence: {confidence}%")
    plt.imshow(image)
    
    
    i +=1
    if i ==9:
        break
```

```python
import numpy as np

class_names = list(test_generator.class_indices.keys())
i=0
res = {
    "pos":0,
    "neg":0
}
for image_batch, label_batch in test_generator:
    
    image = image_batch[0]
    label = int(label_batch[0])
    
    actual_class = class_names[label]
    
    prediction = model.predict(image_batch)
    predicted_class = class_names[np.argmax(prediction[0])]
    if actual_class == predicted_class:
        res["pos"] += 1
    else:
        res["neg"] += 1
    
    i += 1
    if i == 1500:
        break
```
